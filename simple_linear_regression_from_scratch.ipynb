```python
# Importing Necessary Libraries
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
# Size of plot width=20, height=10
plt.rcParams['figure.figsize'] = (20.0, 10.0)
```

```python
# Reading Data
data = pd.read_csv('headbrain.csv')
print(data.shape)
data.head()
```

```python
# Collecting X and Y
# using Head Size to Predict Brain Weight
x = data['Head Size(cm^3)']/100
y = data['Brain Weight(grams)']/100
```

```python
# Plotting Dataset
plt.scatter(x, y, c='#ef5423', label='Scatter Plot')
plt.xlabel('Head Size in cm3')
plt.ylabel('Brain Weight in grams')
plt.legend()
plt.show()
```

```python
# Linear Regression Parameters
learning_rate = 0.0001
iters = 100
m = len(x) # No of Data points

theta = [0.0,0.0]
```

```python
for epoch in range(iters):
    loss = 0
    j0 = 0 # for theta[0]
    j1 = 0 # for theta[1]
    for i in range(m):
        # Compute Hypothesis
        h = theta[0] + theta[1]*x[i]
        # Compute Loss
        loss = h-y[i]
        
        # Gradient Descent derivative term       
        j0 += loss
        j1 += (loss)* x[0]
    
    # Update of Parameters
    theta[0] = theta[0] - learning_rate * (j0/m)
    theta[1] = theta[1] - learning_rate * (j1/m)
    
    # theta value mutiple as 100; as we divided it previously
    print(epoch,theta[0]*100,theta[1]*100)
```

```python
y_pred = []
for i in range(m):
    y_pred.append(theta[0] + theta[1]*x[i])
```

```python
#to plot linear regression line
max_x = np.max(x) 
min_x = np.min(x) 

max_yp = np.max(y_pred) 
min_yp = np.min(y_pred) 

# Calculating line values x and y
x_plt = np.linspace(min_x,max_x,237)
yp = np.linspace(min_yp, max_yp,237)

# Ploting the scores as scatter plot
plt.plot(x_plt, yp, color='#58b970', label='Regression Line')
plt.scatter(x, y, c='#ef5423', label='Scatter Plot')
plt.xlabel('Head Size in cm3')
plt.ylabel('Brain Weight in grams')
plt.legend()
plt.show()
```

```python
# Model Evaluation - RMSE
def rmse(y, y_pred):
    rmse = np.sqrt(sum((y - y_pred) ** 2) / len(y))
    return rmse

# Model Evaluation - R2 Score
def r2_score(y, y_pred):
    mean_y = np.mean(y)
    ss_tot = sum((y - mean_y) ** 2)
    ss_res = sum((y - y_pred) ** 2)
    r2 = 1 - (ss_res / ss_tot)
    return r2
```

```python
print("Mean Square Error : ",rmse(y, y_pred))
print("Accuracy : ",r2_score(y, y_pred))
```
