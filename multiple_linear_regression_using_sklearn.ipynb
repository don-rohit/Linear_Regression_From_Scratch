```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error
#for resize plot
plt.rcParams['figure.figsize'] = (20.0, 10.0)
#for 3d plot
from mpl_toolkits.mplot3d import Axes3D
```

```python
# Reading Data
data = pd.read_csv('student.csv')

print(data.shape)
data.head()
```

```python
math = data['Math']
read = data['Reading']
write = data['Writing']

# Ploting the scores as scatter plot
fig = plt.figure()
ax = Axes3D(fig)
ax.scatter(math, read, write, color='#ef1234',label='Data Point')
ax.set_xlabel('Math')
ax.set_ylabel('Reading')
ax.set_zlabel("Writing")
plt.legend()
plt.show()
```

```python
x = np.array([math,read]).T
y = np.array(write)
```

```python
# Model
model = LinearRegression()
# Data Fitting
reg = model.fit(x,y)
# Y Prediction
y_pred = model.predict(x)
```

```python
# Model Evaluation
rmse = np.sqrt(mean_squared_error(y, y_pred))
accuracy = reg.score(x,y)

print("Mean Square Error : ",rmse)
print("Accuracy : ",accuracy)
```

```python
max_x = np.max(math) + 100
min_x = np.min(math) - 100

max_y = np.max(read) + 100
min_y = np.min(read) - 100

max_yp = np.max(y_pred) + 100
min_yp = np.min(y_pred) - 100

# Calculating line values x and y
x = np.linspace(min_x, max_x, 1000)
y = np.linspace(min_y, max_y, 1000)
yp = np.linspace(min_yp, max_yp, 1000)

# Ploting the scores as scatter plot

fig = plt.figure()
ax = Axes3D(fig)
ax.plot(x, y, yp, color='#58b970', label='Regression Line')
ax.scatter(math, read, write, color='#ef1234',label='Data Point')
ax.set_xlabel('Math')
ax.set_ylabel('Reading')
ax.set_zlabel("Writing")
plt.legend()
plt.show()
```
