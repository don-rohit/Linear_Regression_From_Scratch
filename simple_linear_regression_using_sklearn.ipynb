```python
# Importing Necessary Libraries
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error
# Size of plot width=20, height=10
plt.rcParams['figure.figsize'] = (20.0, 10.0)
```

```python
# Reading Data
data = pd.read_csv('headbrain.csv')
print(data.shape)
data.head()
```

```python
# Collecting X and Y
# using Head Size to Predict Brain Weight
x = data['Head Size(cm^3)'].values/100
y = data['Brain Weight(grams)'].values/100
```

```python
# Plotting Dataset
plt.scatter(x, y, c='#ef5423', label='Scatter Plot')

plt.xlabel('Head Size in cm3')
plt.ylabel('Brain Weight in grams')
plt.legend()
plt.show()
```

```python
# Cannot use Rank 1 matrix in scikit learn
x = x.reshape(len(x), 1)
# Model
model = LinearRegression()
# Data Fitting
reg = model.fit(x,y)
# Y Prediction
y_pred = model.predict(x)
```

```python
# Model Evaluation
rmse = np.sqrt(mean_squared_error(y, y_pred))
accuracy = reg.score(x,y)

print("Mean Square Error : ",rmse)
print("Accuracy : ",accuracy)
```

```python
max_x = np.max(x) 
min_x = np.min(x) 

max_yp = np.max(y_pred) 
min_yp = np.min(y_pred) 

# Calculating line values x and y
x_plt = np.linspace(min_x,max_x,237)
yp = np.linspace(min_yp, max_yp,237)

# Ploting the scores as scatter plot
plt.plot(x_plt, yp, color='#58b970', label='Regression Line')
plt.scatter(x, y, c='#ef5423', label='Scatter Plot')
plt.xlabel('Head Size in cm3')
plt.ylabel('Brain Weight in grams')
plt.legend()
plt.show()
```
